```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv(r'C:/Users/Aarya/Documents/pcos.csv')

# Pie Chart of PCOS Prevalence
pcos_counts = data['PCOS'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(pcos_counts, labels=pcos_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('PCOS Prevalence')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv(r'C:/Users/Aarya/Documents/pcos.csv')

# Stacked Bar Chart for Hair Growth Locations
hair_growth_data = pd.crosstab(data['Hair growth on Chin'], data['PCOS'])

# Create Stacked Bar Chart
hair_growth_data.plot(kind='bar', stacked=True, color=['#ff9999','#66b3ff'], figsize=(8,6))
plt.title('Hair growth on Chin by PCOS Status')
plt.xlabel('Hair growth on Chin')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv(r'C:/Users/Aarya/Documents/pcos.csv')

# Group data by 'Age' and count the occurrences
age_counts = data['Age'].value_counts().sort_index()

# Create Line Plot for Age Distribution
plt.figure(figsize=(8, 6))
sns.lineplot(x=age_counts.index, y=age_counts.values, marker='o', color='b')

# Set titles and labels
plt.title('Age Distribution Line Plot')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.grid(True)

# Display the plot
plt.show()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv(r'C:/Users/Aarya/Documents/pcos.csv')

# Convert 'Yes'/'No' to 1/0
data['Overweight'] = data['Overweight'].map({'Yes': 1, 'No': 0})
data['Difficulty in conceiving'] = data['Difficulty in conceiving'].map({'Yes': 1, 'No': 0})

# Scatter Plot of Overweight vs Difficulty in Conceiving
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='Overweight', y='Difficulty in conceiving', hue='PCOS', palette='Set1')
plt.title('Overweight vs Difficulty in conceiving')
plt.xlabel('Overweight (1 = Yes, 0 = No)')
plt.ylabel('Difficulty in conceiving (1 = Yes, 0 = No)')
plt.show()
```
